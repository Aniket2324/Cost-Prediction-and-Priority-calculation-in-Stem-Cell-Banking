```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
#Loading Dataset
df = pd.read_csv('Dataset.csv')
```

```python
#Adding the mixed column with random points
df['Mixed'] = np.random.randint(1, 10, df.shape[0])
df
```

```python
#calculating rank for various patients w.r.t waiting time
df['Rank_WT']= df['Waiting Time(in months)'].rank(method ='average')
df
```

```python
#defining variables for different factors and taking inputs from the user
print('Enter the values for weightage factor such that sum of Bi is equal to 1')
B1=input('Enter the weightage for Age:')
B2=input('Enter the weightage for Waiting Time:')
B3=input('Enter the weightage for Distance:')
B4=input('Enter the weightage for Urgency:')
B5=input('Enter the weightage for Insured:')
B6=input('Enter the weightage for any other mixed factors:')
```

```python
3#Ai is the point given to patients for different age categories
# Age<20:A1; 20<=Age<40:A2; 40<=Age<60:A3; 60<=Age:A4;
A = [];
for i in range(1,5):
  print('Enter the value of A',i)
  ele=int(input())
  A.append(ele)
print(A)
```

```python
n = len(df.axes[0])
n
```

```python
#Calculating score on the basis of age
ls=[n]
for i in range(0,n):
  if(df['Age'].loc[df.index[i]]<=20):{
    ls.insert(i,A[0])
   }
  elif(20<df['Age'].loc[df.index[i]]<=40):{
    ls.insert(i,A[1])
  }
  elif(40<df['Age'].loc[df.index[i]]<=60):{
    ls.insert(i,A[2])
  }
  elif(df['Age'].loc[df.index[i]]>60):{
      ls.insert(i,A[3])
  }
del ls[10] 
df['Age_Score'] = ls
df
```

```python
1#Similarly for distance(Di)
#D1: Dist<20; D2: 20<=Dist<50; D3: 50<=Dist<100; D4: Dist>100
D = [];
for i in range(1,5):
  print('Enter the value of D',i)
  ele=int(input())
  D.append(ele)
print(D)
```

```python
#Calculating score on the basis of distance
ds=[n]

for i in range(0,n):
  if(df['Distance(in km)'].loc[df.index[i]]<=20):{
    ds.insert(i,D[0])
   }
  elif(20<df['Distance(in km)'].loc[df.index[i]]<=50):{
    ds.insert(i,D[1])
  }
  elif(50<df['Distance(in km)'].loc[df.index[i]]<=100):{
    ds.insert(i,D[2])
  }
  elif(df['Distance(in km)'].loc[df.index[i]]>100):{
      ds.insert(i,D[3])
  }
del ds[n]
df['Dist_Score'] = ds
df
```

```python
#Calculating score on the basis of waiting time
df['Wt_Score'] = df.apply(lambda row: float(B2)*(row.Rank_WT/n), axis = 1)
df
```

```python
#Calculating score on the basis of urgency
df['Urgency_score'] = df.apply(lambda row: float(B4)*(row.Urgency), axis = 1)
df
```

```python
#Calculating score on the basis of insurance
df['Ins_Score'] = df.apply(lambda row: float(B5)*(row.Insured), axis = 1)
df
```

```python
#Calculating score on the basis of mixed factors
df['Mixed_Score'] = df.apply(lambda row: float(B6)*(row.Mixed), axis = 1)
df
```

```python
#Total score of patients
df['Total_Score'] = df.apply(lambda row: float(B1)*(row.Age_Score)+float(B2)*(row.Dist_Score)+(row.Wt_Score)
                                              +(row.Urgency_score)+(row.Ins_Score)+(row.Mixed_Score), axis = 1)
df
```

```python
#calculating the final rank for various patients on the basis of above factors
df['Patient_rank']= n+1-df['Total_Score'].rank(method ='average')
df
```
